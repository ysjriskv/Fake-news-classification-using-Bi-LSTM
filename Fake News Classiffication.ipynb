# TASK #3: PERFORM DATA CLEANING

```python
# download stopwords
nltk.download("stopwords")
```

```python
# Obtain additional stopwords from nltk
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
```

```python
stop_words
```

```python
# Remove stopwords and remove words with 2 or less characters
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
# Apply the function to the dataframe
df['clean'] = df['original'].apply(preprocess)
```

```python
df
```

```python
# Show original news
df['original'][0]
```

```python
# Show cleaned up news after removing stopwords
print(df['clean'][0])
```

```python
# Obtain the total words present in the dataset
list_of_words = []
for i in df.clean:
    for j in i:
        list_of_words.append(j)
```

```python
list_of_words
```

```python
len(list_of_words)
```

```python
# Obtain the total number of unique words
total_words = len(list(set(list_of_words)))
total_words
```

```python
# join the words into a string
df['clean_joined'] = df['clean'].apply(lambda x: " ".join(x))
```

# TASK #5: PREPARE THE DATA BY PERFORMING TOKENIZATION AND PADDING

```python
# split data into test and train 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.clean_joined, df.isfake, test_size = 0.2)
```

```python
from nltk import word_tokenize
```

```python
# Create a tokenizer to tokenize the words and create sequences of tokenized words
tokenizer = Tokenizer(num_words = total_words)
tokenizer.fit_on_texts(x_train)
train_sequences = tokenizer.texts_to_sequences(x_train)
test_sequences = tokenizer.texts_to_sequences(x_test)
```

```python
print("The encoding for document\n",df.clean_joined[0],"\n is : ",train_sequences[0])
```

```python
# Add padding can be maxlen = 4406 or smaller number maxlen = 40
padded_train = pad_sequences(train_sequences,maxlen = maxlen, padding = 'post', truncating = 'post')
padded_test = pad_sequences(test_sequences,maxlen = maxlen,padding = 'post', truncating = 'post') 
```

```python
for i,doc in enumerate(padded_train[:2]):
     print("The padded encoding for document",i+1," is : ",doc)
```

# TASK #6: BUILD AND TRAIN THE MODEL 

```python
# Sequential Model
from tensorflow.keras.layers import Dropout
model = Sequential()

# embeddidng layer
model.add(Embedding(total_words, output_dim = 128))
#model.add(Embedding(total_words, output_dim = 240))
model.add(Dropout(0.3))

# Bi-Directional RNN and LSTM
model.add(Bidirectional(LSTM(128)))
model.add(Dropout(0.3))

# Dense layers
model.add(Dense(128, activation = 'relu'))
model.add(Dense(1,activation= 'sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
total_words
```

```python
y_train = np.asarray(y_train)
```

```python
# train the model
model.fit(padded_train, y_train, batch_size = 64, validation_data = (padded_test, y_test), epochs = 2)
```

# TASK #7: ASSESS TRAINED MODEL PERFORMANCE


```python
# make prediction
pred = model.predict(padded_test)
```

```python
# if the predicted value is >0.5 it is real else it is fake
prediction = []
for i in range(len(pred)):
    if pred[i].item() > 0.5:
        prediction.append(1)
    else:
        prediction.append(0)
```

```python
# getting the accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(list(y_test), prediction)

print("Model Accuracy : ", accuracy)
```

```python
# get the confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(list(y_test), prediction)
plt.figure(figsize = (8, 8))
sns.heatmap(cm, annot = True)
```
